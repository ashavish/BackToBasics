# Plotting Functions in Python

### Visualization Functions in Python

The following functions will help to plot various commonly occuring combinations of continuous and categorical variables for a quick data exploration
<br>
<br>
All functions are available under the class Visualizer. Instantiate an object of this class and then use the methods under it.
<br>
The method **univariate_cont** can be used for univariate analysis of continuous variables.
<br>
The method **univariate_categ** can be used for univariate analysis of categorical variables.
<br>
The method **bivariate_categ_cont** can be used for bivariate analysis of categorical vs continuous variables.
<br>
The method **bivariate_categ_categ** can be used for bivariate analysis of categorical vs categorical variables.
<br>
The method **bivariate_cont_cont** can be used for bivariate analysis of coninuous vs coninuous variables. If a legend is passed of a categorical variable, it can be used for multivariate analysis as well.
<br>
The method **bivariate_contlist_contlist** can be used for multivariate analysis of several continuous variables.
<br>
The method **multivariate_1cont_2categ** can be used for multivariate analysis of 1 continuous and 2 categorical variables.
<br>
We can import a class called **DescriptiveAnalysis** from CommonFunctions.py and use the method **cont_categ_variables** which helps to separate our continuous and categorical variables in a dataset based on the data types provided to it. This should give the right variables only if the data type is set correctly.
<br>
<br>
